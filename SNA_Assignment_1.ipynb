```python
%matplotlib inline
import matplotlib
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import random
import statistics 
```

## Α. Κατασκευή και οπτικοποίηση σύνθετων τύπων δικτύου

### Τοπολογία REG για βαθμό πλέγματος $d = 4$

```python
random.seed(seedno)
n=49 #μέγεθος δικτύου: αντικαταστήστε το x με το τελευταίο ψηφίο του αριθμού μητρώου σας.
d=4
# για την κατασκευή του REG χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
# Επιλέξτε προσεκτικά την τιμή της παραμέτρου p. 
REG=nx.watts_strogatz_graph(49,4,0,seedno) 

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(REG)
```

### Τοπολογία RG (ER) για αριθμό συνδέσεων $M = 150$

```python
random.seed(seedno)
n=49 
M = 150

# για την κατασκευή του RGER χρησιμοποιήστε την gnm_random_graph από το networkx
RGER= nx.gnm_random_graph(49,M,601)

# οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(RGER)
```

### Τοπολογία RGG  με ακτίνα σύνδεσης $R=0.35$

```python
random.seed(seedno)
n=49
R=0.35
pos=0
# για την κατασκευή του RGG χρησιμοποιήστε την random_geometric_graph από το networkx
RGG=nx.random_geometric_graph(49,0.35,dim=2, pos=None, p=2, seed=seedno)

#οπτικοποίηση: https://networkx.org/documentation/stable/auto_examples/drawing/plot_random_geometric_graph.html

# position is stored as node attribute data for random_geometric_graph
pos = nx.get_node_attributes(RGG, "pos")

# find node near center (0.5,0.5)
dmin = 1
ncenter = 0
for n in pos:
    x, y = pos[n]
    d = (x - 0.5) ** 2 + (y - 0.5) ** 2
    if d < dmin:
        ncenter = n
        dmin = d

# color by path length from node near center
p = dict(nx.single_source_shortest_path_length(RGG, ncenter))

plt.figure(figsize=(8, 8))
nx.draw_networkx_edges(RGG, pos, alpha=0.4)
nx.draw_networkx_nodes(
    RGG,
    pos,
    nodelist=list(p.keys()),
    node_size=80,
    node_color=list(p.values()),
    cmap=plt.cm.Reds_r,
)

plt.xlim(-0.05, 1.05)
plt.ylim(-0.05, 1.05)
plt.axis("off")
plt.show()
```

### Τοπολογία SF με παράμετρο επισύναψης $d=4$

```python
random.seed(seedno)
n=49 
d=4

# για την κατασκευή του SF χρησιμοποιήστε την barabasi_albert_graph από το networkx
SF=nx.barabasi_albert_graph(49,4,seedno) 

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(SF)
```

### Τοπολογία SW με βαθμό αρχικού πλέγματος $d=4$ και πιθανότητα ανασύνδεσης $p=0.25$

```python
random.seed(seedno)
n=49
d=4
p=0.25

# για την κατασκευή του SW χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
SW=nx.watts_strogatz_graph(49,d,p,seedno) 

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(SW)
```

## B. Μετρικές δικτύου

### B1. Συντελεστής ομαδοποίησης (Clustering coefficient)

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β1.1 Να υπολογιστεί ο συντελεστής ομαδοποίησης κάθε κόμβου και να αναπαρασταθεί η κατανομή του.
#### Β1.2  Να υπολογιστεί ο μέσος συντελεστής ομαδοποίησης και η διασπορά του.
#### Β.1.3 Να βρεθούν οι κόμβοι με το μικρότερο και το μεγαλύτερο συντελεστή ομαδοποίησης. 
#### Να σχολιαστούν τα αποτελέσματα.

```python
def cc(G):
    return nx.clustering(G)

#print(cc(REG),"\n")
#print(cc(RGER),"\n")
#print(cc(SF),"\n")
#print(cc(SW),"\n")
REGcc=cc(REG)
RGERcc=cc(RGER)
RGGcc=cc(RGG)
SFcc=cc(SF)
SWcc=cc(SW)
array1=[REG,RGER,RGG,SF,SW]
array2=["REG","RGER","RGG","SF","SW"]
array=[REGcc,RGERcc,RGGcc,SFcc,SWcc]
for i in range(0,len(array)):
    print(array[i])
    arr_keys=list(array[i].keys())
    arr_values=list(array[i].values())
    plt.hist(arr_values)
    plt.show()
    
average=[] 


for i in range(0,len(array1)):
    a=nx.average_clustering(array1[i])
    average.append(a)
    print(array2[i],'average clustering:',a,'\n')


for i in range(0,len(array)):
    b=statistics.variance(array[i].values(),xbar=average[i])
    print(array2[i],'cc variance:',b,'\n')

for i in range(0,len(array)):
    all_values=list(array[i].values())
    print(array2[i],"max cc",':',max(all_values),"at",all_values.index(max(all_values)),'\n')
    print(array2[i],"min cc",':',min(all_values),"at",all_values.index(min(all_values)),'\n')
               
               
```

#### Παρατηρούμε πως στο γράφο REG όλοι οι κόμβοι έχουν όμοιο clustering coefficient, παράλληλα στους υπόλοιπους γράφους, παρατηρούμε πως κάθε κόμβος παίρνει διάφορες τιμές clustering coefficient. Ο γράφος με τη μέγιστη μέση τιμή ομαδοποίησης είναι ο RGG και εκείνος με την ελάχιστη ο RGER. Ο γράφος με την ελάχιστη τιμή διασποράς της ομαδοποιήσης είναι ο REG ενώ εκείνος με τη μέγιστη είναι ο SW. Οι γράφοι που περιέχουν κόμβους με τη μεγιστη τιμή που μπορεί να έχει το clustering coefficient (1.0) είναι οι SW και RGG, ενώ εκείνοι οι γράφοι που περιέχουν έστω και έναν κόμβο με μηδενικό clustering coefficient είναι οι RGER, SF, SW.  

### Β2. Μέσο μήκος μονοπατιού - Διάμετρος - Ακτίνα γράφου

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β2.1 Nα υπολογιστεί το μέσο μήκος μονοπατιού και η διασπορά του. Προτείνεται η χρήση της συνάρτησης  all_pairs_shortest_path_length του networkx.
#### B2.2 Να υπολογιστεί η διάμετρος (συνάρτηση diameter) και η ακτίνα (συνάρτηση radius).
#### Για τα παραπάνω πραγματοποιήστε συγκρίσεις και εξηγήστε με βάση τα χαρακτηριστικά της κάθε τοπολογίας όπως παρουσιάζονται στη θεωρία.

```python
def compute_path_lengths(G):
    return dict(nx.all_pairs_shortest_path_length(G))
all=[]
variance=[]
average_path_lengths=[]
for i in range(0,len(array)):
    paths=compute_path_lengths(array1[i])
    k=0
    for j in range(0,len(paths)):        
        c=paths[j].values()
        paths[j]=sum(c)/(len(c)-1)
        k=sum(c)+k
    var=statistics.variance(paths.values())
    variance.append(var)
    a=k/(len(paths)*(len(paths)-1))
    average_path_lengths.append(a)
    all.append(paths)
    print(array2[i],"average path length :",a,"\n")
    print(array2[i],"average path length using networkx function:",nx.average_shortest_path_length(array1[i]),"\n")
    print(array2[i],"variance of average path length :",var,"\n")
 
for i in range(0,len(array)):
    d=nx.diameter(array1[i])
    r=nx.radius(array1[i])
    print(array2[i],"diameter :",d,"\n")
    print(array2[i],"radius :",r,"\n")

```

####  Το μέσο μήκος μονοπατιού για το γράφο REG είναι σταθερό και σύμφωνα με τη θεωρία περιμένουμε πως τα μέσα μήκη για τους γράφους SW και SF θα θεωρούνται σχετικά μικρά. Τα μέσα μήκη για τους γράφους RGG και RGER σύμφωνα με τη θεωρία περιμένουμε πως θα είναι σχετικά μεγάλα στην πρώτη περίπτωση και σχετικά μέτρια στη δεύτερη. Στη συγκεκριμένη περίπτωση παρατηρούμε πως ως προς τη μέση τιμή του μήκους μονοπατιού με αύξουσα σειρά οι τοπολογίες θα κατατασσονται ως εξής:
    
#####  SF, RGG, RGER, SW, REG

#### Αναφορικά με τη διασπορά σε κάθε περίπτωση θα έχουμε πάλι με αύξουσα σειρά:

#####   REG, SF, SW, RGER, RGG

#### Τέλος βλέπουμε πως τη μεγαλύτερη διάμετρο έχει η τοπολογία REG και ακολουθούν οι τοπολογίες SF, RGER, RGG, SW. Επίσης σχετικά με την ακτίνα παρατηρούμε πως η τοπολογία REG έχει την μεγαλύτερη τιμή (ισούται και με την τιμή της διαμέτρου) και ακολουθούν οι τοπολογίες SW, RGER, RGG, SW. Θα μπορούσαμε να βρούμε την ακτίνα και τη διάμετρο βρίσκοντας την ελάχιστη και τη μέγιστη τιμή της εκκεντρότητας αντίστοιχα σε κάθε τοπολογία.


#### B.2.3 Για κάθε τοπολογία, ποιος κόμβος συνδέεται με τους περισσότερους κόμβους με μήκος ελάχιστου μονοπατιού ίσο με τη διάμετρο της τοπολογίας?

```python

def find_nodes(G):
    dictionary={}
    temp=[]
    nodes=[]
    labels=[]
    paths=compute_path_lengths(G)
    connectivity_node=0
    for j in range(0,len(paths)):
        c=paths[j].values()
        v=paths[j].keys()
        c_list=list(c)
        v_list=list(v)
        temp=[]
        for k in range(0,len(c)):
            if(c_list[k]==nx.diameter(G)):
                temp.append(v_list[k])
        if(len(temp)>len(nodes)):
            dictionary={}
            nodes=[]
            nodes=temp
            labels=[]
            labels.append(j)
            dictionary[j]=nodes
        elif(len(temp)==len(nodes)):
            nodes=temp
            dictionary[j]=nodes

    return dictionary
    
for i in range(0,len(array)):
    print(array2[i],"{labels : node set}",":",find_nodes(array1[i]),"\n")
#Σε περίπτωση ισότητας πλήθους συνδέσεων δίνουμε όλους τους κόμβους και όλα τ' αντίστοιχα σύνολα
```

#### Περιφέρεια του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με τη διάμετρό του. 
#### Κέντρο του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με την ακτίνα του.  
#### Β.2.4 Για κάθε τοπολογία, να βρεθεί η περιφέρεια και το κέντρο της και να σχολιαστούν τα αποτελέσματα.

```python
def find_periphery(G):
    #dictionary={}
    periphery=[]
    #center=[]
    b=nx.eccentricity(G)
    b_keys=list(b.keys())
    b_values=list(b.values())
    #r=nx.radius(G)
    d=nx.diameter(G)
    for i in range(0,len(b)):
        if(b_values[i]==d):
            periphery.append(b_keys[i])
    return periphery

for i in range(0,len(array)):
    print(array2[i],"periphery:",find_periphery(array1[i]),"\n")
```

#### Παρατηρούμε πως όλοι οι κόμβοι στο γράφους REG και SF αποτελούν την περιφέρεια. Στη συνέχεια βλέπουμε πως οι γράφοι RGER και RGG έχουν περιφέρεια που αποτελείται από τέσσερις κόμβους και ο γράφος SW έχει περιφέρεια που αποτελείται από έντεκα κόμβους.  

```python
def find_center(G):
    #dictionary={}
    #periphery=[]
    center=[]
    b=nx.eccentricity(G)
    b_keys=list(b.keys())
    b_values=list(b.values())
    r=nx.radius(G)
    #d=nx.diameter(G)
    for i in range(0,len(b)):
        if(b_values[i]==r):
            center.append(b_keys[i])
    return center

for i in range(0,len(array)):
    print(array2[i],"center:",find_center(array1[i]),"\n")   
```

#### Παρατηρούμε πως το κέντρο του REG αποτελείται από όλους τους κόμβους του γράφου. Οι γράφοι RGER και RGG έχουν κέντρα που αποτελούνται από έντεκα κόμβους. Ο γράφος SF έχει κέντρο που αποτελείται από δύο κόμβους και ο γράφος SW έχει κέντρο που αποτελείται από τέσσερις κόμβους.

## Γ) Mετρικές κεντρικότητας κόμβων

### Γ1. Κεντρικότητα βαθμού (Degree centrality)
##### Γ1.1 Να αναπαρασταθεί η κατατανομή βαθμών κόμβων για κάθε τοπολογία.
##### Γ1.2 Να υπολογιστεί ο μέσος βαθμός κόμβου καθώς και η διασπορά των βαθμών κόμβου για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_degree(G):
    return nx.degree_centrality(G)

dc=[]

for i in range(0,len(array)):
    dc.append(find_degree(array1[i]))

for i in range(0,len(array)):
    plt.hist(dc[i].values())
    plt.show()

for i in range(0,len(array)):
    a=statistics.mean(dc[i].values())
    b=statistics.variance(dc[i].values())
    print(array2[i]," mean degree centrality :",a,"\n")
    print(array2[i],"variance of degree centrality :",b,"\n")
```

#### Γνωρίζουμε από τη θεωρία πως η κατανομή του βαθμού για τοπολογία REG θα είναι dirac, για SW heavy-tailed , για SF power law, για RGER περιμένουμε Poisson και για RGG uniform. Σε σχέση με το μέσο βαθμού κόμβου βλέπουμε πως τη μικρότερη τιμή λαμβάνουν οι τοπολογίες SW, REG και στη συνέχεια ακολουθούν με αύξουσα σειρά οι τοπολογίες RGER, SF, RGG. Στη συνέχεια κατατάσσοντας σε αύξουσα σειρά τις τοπολογίες με βάση τη διασπορά του βαθμού, εκείνος ο κόμβος με τη μικρότερη διασπορά θα είναι ο REG και ακολουθούν οι  SW, RGER, SF και RGG.

### Γ2. Κεντρικότητα εγγύτητας (Closeness centrality)
##### Γ2.1 Να αναπαρασταθεί η κατατανομή της εγγύτητας κόμβων για κάθε τοπολογία.
##### Γ2.2 Να υπολογιστεί η μέση εγγύτητα καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_closeness(G):
    return nx.closeness_centrality(G)

cce=[]

for i in range(0,len(array)):
    cce.append(find_closeness(array1[i]))

for i in range(0,len(array)):
    plt.hist(cce[i].values())
    plt.show()

for i in range(0,len(array)):
    c=statistics.mean(cce[i].values())
    d=statistics.variance(cce[i].values())
    print(array2[i]," mean closeness centrality :",c,"\n")
    print(array2[i],"variance of closeness centrality :",d,"\n")
```

#### Παρατηρούμε πως η τοπολογία REG έχει τη μικρότερη μέση τιμή closeness centrality και στη συνέχεια με αύξουσα σειρά ακολουθούν οι τοπολογίες SW, RGER, RGG και SF. Σχετικά με τη διασπορά του συνόλου τιμών closeness centrality για κάθε διαφορετική τοπολογία βλέπουμε πως η τοπολογία REG όπως αναμέναμε έχει την μικρότερη τιμή και ακολουθούν με άυξουσα σειρά οι τοπολογίες SW, RGER, SF, RGG. 

### Γ3. Ενδιαμεσική κεντρικότητα (Betweenness centrality)
##### Γ3.1 Να αναπαρασταθεί η κατατανομή της ενδιαμεσικής κεντρικότητας κόμβων για κάθε τοπολογία.
##### Γ3.2 Να υπολογιστεί η μέση ενδιαμεσική κεντρικότητα καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_betweenness(G):
    a=nx.betweenness_centrality(G,k=None)
    return a

bce=[]
for i in range(0,len(array)):
    bce.append(find_betweenness(array1[i]))
  
for i in range(0,len(array)):
    print(bce[i])
    plt.hist(bce[i].values())
    plt.show()

for i in range(0,len(array)):
    e=statistics.mean(bce[i].values())
    f=statistics.variance(bce[i].values())
    print(array2[i]," mean closeness centrality :",e,"\n")
    print(array2[i],"variance of closeness centrality :",f,"\n")
    
```

#### Για κάποιο λόγο -που δεν κατάφερα να εντοπίσω- το διάγραμμα για το γράφο REG δεν σχηματίζεται, όμως φαίνεται από τις τιμές που επιστρέφονται πως για όλους τους κόμβους θα είναι ίδια η τιμή betweenness centrality όπως αναμένουμε. Ανάμεσα στις τοπολογίες που συναντούμε τη μεγαλύτερη μέση τιμή betweenness centrality παρουσιάζει ο γράφος REG και ακολουθουν με φθίνουσα σειρά οι SW, RGER, RGG και SF. Αναμενόμενα όπως και σ' όλα τα υπόλοιπα ερωτήματα η διασπορά κάθε είδους centrality για τον REG θα είναι 0 και στη συνέχεια με αύξουσα σειρά ακολουθούν οι RGER, RGG, SW και SF.

### Γ4. Kεντρικότητα Katz
##### Γ4.1 Να αναπαρασταθεί η κατανομή της κεντρικότητας Katz των κόμβων για κάθε τοπολογία. 
Για τον προσδιορισμό της παραμέτρου $alpha$ της συνάρτησης katz_centrality,  να υπολογιστεί η μεγαλύτερη ιδιοτιμή $λ_{max}$ του πίνακα γειτνίασης με τη συνάρτηση adjacency_spectrum του networkx. 
H $alpha$ ορίζεται να είναι ιση με $\frac{1}{λ_{max}}-0.01$.
##### Γ4.2 Να υπολογιστεί η μέση  κεντρικότητα Katz  καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_katz(G):
    lambda_max=max(nx.adjacency_spectrum(RGG, weight='weight'))
    alpha=(1/(lambda_max.real))-0.01
    return nx.katz_centrality(G, alpha)

kc=[]

for i in range(0,len(array)):
    kc.append(find_katz(array1[i]))

for i in range(0,len(array)):
    plt.hist(kc[i].values())
    plt.show()    
    
for i in range(0,len(array)):
    g=statistics.mean(kc[i].values())
    h=statistics.variance(kc[i].values())
    print(array2[i]," mean Katz centrality :",g,"\n")
    print(array2[i],"variance of Katz centrality :",h,"\n")
```

####  Παρατηρούμε πως όλες οι μέσες τιμές της κεντρικότητας είναι ίδιας τάξης και δεν έχουμε μεγάλες αποκλίσεις μεταξύ των διαφόρων ειδών τοπολογιών. Όλες οι διασπορές είναι πολύ μικρές όμως ο γράφος με τη μεγαλύτερη διασπορά είναι ο RGG εκείνος με τη μικρότερη είναι ο SW εννοείται μετά τον REG που σε κάθε περίπτωση έχει μηδέν αφού οι τιμές δεν μεταβάλλονται.

#### Να προσδιοριστούν, για κάθε τοπολογία, οι πιο κεντρικοί κόμβοι ως προς το βαθμό, την εγγύτητα, την ενδιαμεσική κεντρικότητα και την κεντρικότητα Katz. Να σχολιαστούν τα αποτελέσματα.

```python
def find_central(G):
    tempA=[]
    tempB=[]
    tempC=[]
    tempD=[]
    dictionary={}
    a=find_degree(G)
    b=find_closeness(G)
    c=find_betweenness(G)
    d=find_katz(G)
    key_values=list(a.keys())
    a_values=list(a.values())
    maxA=max(a_values)
    b_values=list(b.values())
    maxB=max(b_values)
    c_values=list(c.values())
    maxC=max(c_values)
    d_values=list(d.values())
    maxD=max(d_values)
    for value in range(0,len(a)):
        if(a[value]==maxA):
            tempA.append(value)
    for value1 in range(0,len(b)):
        if(b[value1]==maxB):
            tempB.append(value1)
    for value2 in range(0,len(c)):
        if(c[value2]==maxC):
            tempC.append(value2)
    for value3 in range(0,len(d)):
        if(d[value3]==maxD):
            tempD.append(value3)
            
    dictionary["degree centrality"]=tempA
    dictionary["closeness centrality"]=tempB
    dictionary["betweenness centrality"]=tempC
    dictionary["Katz centrality"]=tempD
    return dictionary

for i in range(0,len(array)):
    a=find_central(array1[i])
    print(array2[i]," nodes per category :",a,"\n")
```

Στους γράφους SF και SW παρατηρούμε πως ο κόμβος με τη μέγιστη κεντρικότητα βαθμού σε κάθε περίπτωση αποτελεί την απάντηση και στα υπόλοιπα είδη κεντρικότητας.
